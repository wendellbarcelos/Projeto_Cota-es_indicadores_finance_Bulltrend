### Desafio 1

Extrair os dados de cotações diárias dos ativos no período de 01/01/2023 até 31/08/2023.

Dica: Extraia a data, os dados OHLC (Open, High, Low, Close) e o ticker do Ativo como colunas da tabela cotacoes.

```python
# Carregando as bibliotecas

import yfinance as yf
import pandas as pd
```

```python
# Gerando uma carteira com os ativos

carteira_yf = ['ASAI3.SA','EQTL3.SA','LREN3.SA','PRIO3.SA', 'RAIL3.SA', 'RDOR3.SA', 'SBSP3.SA',	'VIVT3.SA']
```

```python
# Puxando os dados da API

df = yf.download(carteira_yf, start='2023-01-01', end='2024-01-01')
```

```python
# Visualizando os dados
df.head(5)
```

```python
#Passando os ativos para o multindex do df
cotacoes = df.stack(level=1)
cotacoes
```

```python
# Alterando o indice e renomeando colunas

cotacoes = cotacoes.reset_index().rename(columns= {'Ticker':'Ativos'})
cotacoes
```

```python
# Retornando as colunas Ativos, Datas e OHLC

cotacoes = cotacoes[['Ativos', 'Date', 'Open', 'High', 'Low', 'Close']]
cotacoes
```
